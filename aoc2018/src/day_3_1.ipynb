```python
import pandas as pd
df=pd.DataFrame(columns=["timestamp", "text"])
f = open("input_day_4.txt", "r")
for line in f:
    pd_input = line.split("]")
    pd_input[0] = pd_input[0][1:]
    pd_input[1] = pd_input[1][1:-1]
    df = df.append({"timestamp": pd_input[0], "text" : pd_input[1]}, ignore_index=True)
print(df)

```

```python
df_sorted=df.sort_values(by=['timestamp'])
print(df_sorted)
```

```python
df_sorted['minutes']=df_sorted['timestamp'].apply(lambda x: int(x[14:]) )
print(df_sorted)
```

```python
#df_sorted = df_sorted.drop(columns=["timestamp"])
df_sorted.to_csv(r'clean_output.txt', header=None, index=None, sep=',', mode='a')
df_sorted.head()
    
```
